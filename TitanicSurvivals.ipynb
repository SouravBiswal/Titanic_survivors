```python
import pandas as pd
```

```python
df= pd.read_csv('titanic.csv')
```

```python
df.head(50)
```

```python
df_new= df.drop(['Name','PassengerId','Ticket','Parch','Cabin','Embarked','SibSp'],axis='columns')
```

Did some manipulations on the dataset to perform analysis productively

```python
df_new.head(25)
```

```python
import numpy as np
```

```python
Avg_age=df_new.Age.median()
Avg_age
```

```python
df_new.Age=df_new.Age.fillna(Avg_age)
df_new.head(25)
```

```python
inputs= df_new.drop('Survived',axis='columns')
target= df_new['Survived']
```

```python
target.head(5)
```

```python
inputs.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Encoded_Sex= LabelEncoder()
```

```python
inputs['Gender']= Encoded_Sex.fit_transform(inputs['Sex'])
inputs.head(5)
```

```python
inputs_rev=inputs.drop('Sex',axis='columns')
```

```python
from sklearn import tree
```

```python
model= tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(inputs_rev,target,test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python
y_test.head
```

```python
X_test
```

```python
model.predict(X_test)
```

```python
model.score(inputs_rev,target)
```

Score is 0.9416, which means that accuracy of our trained model is 94.16%.
